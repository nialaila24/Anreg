Pengklasifikasian Decision Tree di Scikit-Learn

Import Library

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
```

Menginput data "heart"

```python
col_names = ['age','sex','cp','trtbps','chol','fbs','restecg','thalachh','exng','oldpeak','slp','caa','thall','label']
df =pd.read_csv('/content/heart.csv')
df.head()
```

```python
# mengetahui jumlah baris dan kolom
print("Jumlah baris: %d" % df.shape[0])
print("Jumlah kolom: %d" % df.shape[1])
```

Seleksi fitur

```python
features = ['age','sex','cp','trtbps','chol','fbs','restecg','thalachh','exng','oldpeak','slp','caa','thall']
X = df[features]
y = df['output']
print(X)
print(y)
```

Memisahkan data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training dan 30% test
```

Membangun Decision Tree

```python
#Membuat Objek Decision Tree classifer 
clf = DecisionTreeClassifier()

# Pengklasifikasi Decision Tree 
clf = clf.fit(X_train,y_train)

#Predict respon untuk test dataset
y_pred = clf.predict(X_test)
```

Mengevaluasi Model

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Pengoptimalan Kinerja Decision Tree

```python
# Membuat objek Decision Tree classifer 
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Pengklasifikasi Decision Tree 
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
